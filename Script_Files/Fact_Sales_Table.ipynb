### Sales Table 

##### Import Packages

```python
import pandas as pd
import numpy as np
import csv
import random
from faker import Faker
```

```python
# Mention the number of rows
num_rows = int(input("Enter the number of row the CSV file should have: "))

# generate series of dates
date_series = np.arange(np.datetime64('2010-01-01'),np.datetime64('2025-05-01'))
random_dates = np.random.choice(date_series, size=num_rows)

formatted_row = pd.to_datetime(random_dates).strftime('%Y%m%d')

data = {
    "Customer_ID" : np.random.randint(1,1001,size=num_rows),
    "Store_ID" : np.random.randint(1,101,size=num_rows),
    "Order_Date_ID" : formatted_row,
    "Product_ID" : np.random.randint(1,1001,size=num_rows),
    "Quantity_Ordered" : np.random.randint(1,21,size=num_rows),
    "Order_Amount" : np.random.randint(100,1001,size=num_rows)  
}

sales_df = pd.DataFrame(data)

# creating percentage columns
discount_percent = np.random.uniform(0.02, 0.15, size = num_rows)
shipping_percent = np.random.uniform(0.05, 0.15, size = num_rows)

# Including percentage and total columns in sales table
sales_df['Discount_Amount'] = round(sales_df["Order_Amount"] * discount_percent,2)
sales_df['Shipping_Cost'] = round(sales_df["Order_Amount"] * shipping_percent,2)
sales_df['Total_Amount'] = sales_df["Order_Amount"] - (sales_df['Discount_Amount'] + sales_df['Shipping_Cost'])

sales_df.to_csv("Fact_Orders_Table.csv",index=False)

# Print Success
print("File created successfully")
sales_df
```
